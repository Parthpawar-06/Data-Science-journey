```python
import numpy as np
```

```python
data = [10, 20, 20, 30, 40, 1000]
```

```python
print(np.mean(data))      #calculating mean
print(np.median(data))    #calculating median
```

```python
data = [10, 12, 11, 13, 10, 14]
```

```python
print(np.var(data))       #calculating variance
print(np.std(data))       #calculating standard deviation
```

```python
import matplotlib.pyplot as plt
data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data, bins=30)
plt.title("Normal distribution — mean=0, std=1")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset: salaries of 10 employees
salaries = [35000, 42000, 38000, 55000, 61000,
            48000, 52000, 39000, 200000, 44000]

print("Mean:  ", round(np.mean(salaries), 2))
print("Median:", round(np.median(salaries), 2))
print("Std Dev:", round(np.std(salaries), 2))
print("Variance:", round(np.var(salaries), 2))
print("Min:", min(salaries), "Max:", max(salaries))

# Question to answer: why is mean much higher than median here?
# Answer: 200000 is an outlier pulling the mean up.
# This is why median is used for salary reporting, not mean.

plt.hist(salaries, bins=8, color='steelblue', edgecolor='white')
plt.axvline(np.mean(salaries), color='red', label='Mean')
plt.axvline(np.median(salaries), color='green', label='Median')
plt.legend()
plt.title("Salary distribution — notice the outlier effect")
plt.show()
```
